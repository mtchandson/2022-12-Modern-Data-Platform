```
import pyspark
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("parquetFile").getOrCreate()
data =[("James ","","Smith","36636","M",3000),
              ("Michael ","Rose","","40288","M",4000),
              ("Robert ","","Williams","42114","M",4000),
              ("Maria ","Anne","Jones","39192","F",4000),
              ("Jen","Mary","Brown","","F",-1)]
columns=["firstname","middlename","lastname","dob","gender","salary"]
df=spark.createDataFrame(data,columns)
df.write.mode("overwrite").parquet("dbfs:/mnt/prepared/output/people.parquet")

```

```

parDF1=spark.read.parquet("dbfs:/mnt/prepared/output/people.parquet")

parDF1.createOrReplaceTempView("parquetTable")
parDF1.printSchema()
parDF1.show(truncate=False)

```

```

parkSQL = spark.sql("select * from ParquetTable where salary >= 4000 ")
parkSQL.show(truncate=False)

spark.sql("CREATE TEMPORARY VIEW PERSON11 USING parquet OPTIONS (path \"dbfs:/mnt/prepared/output/people.parquet\")")
spark.sql("SELECT * FROM PERSON11").show()

df.write.partitionBy("gender","salary").mode("overwrite").parquet("dbfs:/mnt/prepared/output/people2.parquet")

parDF2=spark.read.parquet("dbfs:/mnt/prepared/output/people2.parquet/gender=M")
parDF2.show(truncate=False)

spark.sql("CREATE TEMPORARY VIEW PERSON2 USING parquet OPTIONS (path \"dbfs:/mnt/prepared/output/people2.parquet/gender=F\")")
spark.sql("SELECT * FROM PERSON2" ).show()
```
